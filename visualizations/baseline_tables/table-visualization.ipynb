```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
KY_baseline_data = pd.read_csv('ky_baseline.csv')
KY_interpretable_data = pd.read_csv('ky_interpretable.csv')
FL_baseline_data = pd.read_csv('fl_baseline.csv')
FL_interpretable_data = pd.read_csv('fl_interpretable.csv')
```

## KY Baseline

```python
KY_baseline_data_two_year = KY_baseline_data[KY_baseline_data['Time'] == 'Two Year']
KY_baseline_data_six_month = KY_baseline_data[KY_baseline_data['Time'] == 'Six Month']
```

```python
plt.style.use('ggplot')
fig, axs = plt.subplots(ncols=2, figsize=(14,5))
fig.suptitle('Kentucky Baseline Models', fontsize = 18, y = 1)
sns.barplot(x='Problem', y='Performance', hue='Model', data=KY_baseline_data_two_year, ax=axs[0])
axs[0].set_title('Two Year', fontsize = 16)
axs[0].get_legend().remove()
axs[0].set_xlabel('Problem', fontsize=16)
axs[0].set_ylabel('Performance - avg. AUC', fontsize=16)

sns.barplot(x='Problem', y='Performance', hue='Model', data=KY_baseline_data_six_month, ax=axs[1])
axs[1].set_title('Six Month', fontsize = 16)
axs[1].legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
axs[1].set_xlabel('Problem', fontsize=16)
axs[1].set_ylabel("")
pass
```

## KY Interpretable

```python
KY_interpretable_data_two_year = KY_interpretable_data[KY_interpretable_data['Time'] == 'Two Year']
KY_interpretable_data_six_month = KY_interpretable_data[KY_interpretable_data['Time'] == 'Six Month']
```

```python
plt.style.use('ggplot')
fig, axs = plt.subplots(ncols=2, figsize=(14,5))
fig.suptitle('Kentucky Interpretable Models', fontsize = 15, y = 1)
sns.barplot(x='Problem', y='Performance', hue='Model', data=KY_interpretable_data_two_year, ax=axs[0])
axs[0].set_title('Two Year', fontsize = 14)
axs[0].legend("")
axs[0].set_xlabel('Problem', fontsize=13)
axs[0].set_ylabel('Performance - avg. AUC', fontsize=13)
sns.barplot(x='Problem', y='Performance', hue='Model', data=KY_interpretable_data_six_month, ax=axs[1])
axs[1].set_title('Six Month', fontsize = 14)
axs[1].legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
axs[1].set_xlabel('Problem', fontsize=13)
axs[1].set_ylabel("")
pass
```
